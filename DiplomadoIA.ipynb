Las decisiones se tomas desde la estadistica Inferencial, no desde la descriptiva.

📊 Estadística descriptiva
Qué hace: Resume y presenta los datos disponibles. Usa medidas como promedios, medianas, desviaciones estándar, gráficos, etc.
La estadística descriptiva se encarga de recoger, organizar, resumir y presentar datos de manera clara y comprensible. No busca sacar conclusiones más allá de los datos observados, sino simplemente describir sus características principales.

Ejemplo: “El 60% de los pacientes en este estudio tiene hipertensión.”

Alcance: Solo describe los datos de la muestra observada, sin generalizar.

🔍 # Estadística inferencial
Qué hace: Permite tomar decisiones o sacar conclusiones sobre una población completa, a partir de una muestra. Utiliza técnicas como pruebas de hipótesis, intervalos de confianza, regresión, etc.

Ejemplo: “Con un 95% de confianza, estimamos que entre el 55% y 65% de toda la población tiene hipertensión.”

Alcance: Va más allá de los datos observados para predecir o decidir sobre lo que no se ha observado.

🎯 Entonces, ¿por qué se toman decisiones con estadística inferencial?
Porque en la mayoría de los casos:

No tenemos acceso a toda la población, solo a una muestra.

Queremos aplicar nuestras conclusiones más allá de los datos analizados, como en medicina, negocios, ciencia, etc.

Necesitamos cuantificar el nivel de incertidumbre de nuestras conclusiones (por ejemplo, ¿con qué nivel de confianza podemos decir que un nuevo medicamento funciona mejor que el anterior?).

La estadística descriptiva te dice lo que pasa en tus datos. La inferencial te permite tomar decisiones sobre lo que no has observado directamente.

### 🧠 ¿Qué es un **test de hipótesis**?

Es un procedimiento estadístico que nos ayuda a **tomar decisiones** o **sacar conclusiones** sobre una **población**, a partir de los datos de una **muestra**.

---

### ⚖️ ¿Cómo funciona?

El test de hipótesis compara dos afirmaciones:

1. **Hipótesis nula (H₀)**: Es la afirmación que queremos **probar o rechazar**. Representa el “estado actual” o la “no diferencia”.
2. **Hipótesis alternativa (H₁ o Hₐ)**: Es lo que queremos demostrar, o lo que creemos que puede estar ocurriendo en realidad.

---

### 📌 Ejemplo sencillo

**Situación:** Una fábrica afirma que sus bombillas duran en promedio 1,000 horas. Queremos saber si realmente eso es cierto.

---

#### 1. **Planteamos las hipótesis**

* **H₀**: μ = 1000  → la duración media es 1,000 horas
* **H₁**: μ ≠ 1000  → la duración media **no** es 1,000 horas

*(Esto se llama test bilateral)*

---

#### 2. **Tomamos una muestra**

Supón que tomamos 50 bombillas y calculamos su media (por ejemplo, 980 horas) y su desviación estándar.

---

#### 3. **Calculamos el estadístico de prueba**

Aquí usamos una fórmula (por ejemplo, un **t de Student** o un **z** si conocemos la desviación poblacional).

$$
z = \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}}
$$

---

#### 4. **Fijamos el nivel de significancia α**

Típico:

* α = 0.05 → 5% de tolerancia para cometer un error tipo I (rechazar H₀ cuando era verdadera).

---

#### 5. **Decisión**

* Si el valor p (p-value) es **menor que α**, **rechazamos H₀** → hay evidencia estadística de que **la duración no es 1,000 horas**.
* Si el valor p es **mayor que α**, **no rechazamos H₀** → no tenemos suficiente evidencia para decir que la duración sea distinta de 1,000 horas.

---

### 🧾 Tipos de errores

| Tipo              | Qué es                       | Consecuencia   |
| ----------------- | ---------------------------- | -------------- |
| **Error Tipo I**  | Rechazar H₀ siendo verdadera | Falso positivo |
| **Error Tipo II** | No rechazar H₀ siendo falsa  | Falso negativo |

Vamos a resolver un **test de hipótesis paso a paso** con números reales.

---

### 🧪 **Problema**:

Una empresa dice que el contenido promedio de sus botellas de agua es de **500 ml**. Queremos saber si eso es cierto.

---

### 🔢 Datos de la muestra:

* Tamaño de muestra: n = 30 botellas
* Media muestral: $\bar{x} = 492$ ml
* Desviación estándar muestral: $s = 10$ ml
* Nivel de significancia: $\alpha = 0.05$
* Hipótesis:

  * H₀: μ = 500 ml
  * H₁: μ ≠ 500 ml *(test bilateral)*

---

### 📐 Paso 1: Calcular el estadístico de prueba (t de Student)

Usamos esta fórmula:

$$
t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}} = \frac{492 - 500}{10/\sqrt{30}} = \frac{-8}{10/\sqrt{30}} \approx \frac{-8}{1.8257} \approx -4.38
$$

---

### 📊 Paso 2: Determinar el valor crítico o p-value

* Usamos la **distribución t** con **n - 1 = 29 grados de libertad**
* Con α = 0.05 (bilateral), el **valor crítico** para 29 g.l. es aproximadamente **±2.045**

---

### ✅ Paso 3: Regla de decisión

* Si $t < -2.045$ o $t > 2.045$, **rechazamos H₀**
* Nuestro $t = -4.38$, que es **mucho menor** que -2.045

---

### 🧾 Paso 4: Conclusión

> Como el valor de $t$ cae en la **región de rechazo**, **rechazamos la hipótesis nula H₀**.
> Es decir, hay evidencia estadísticamente significativa de que el promedio **no es 500 ml**.

### 🎯 Contexto:

Cuando realizamos un test de hipótesis, siempre hay **incertidumbre**, ya que trabajamos con **muestras** y no con toda la población. Por eso, podemos cometer dos tipos de errores:

---

## 🔴 **Error Tipo I (α)**

> **Rechazar la hipótesis nula (H₀) siendo verdadera.**

### 🧠 Interpretación:

* Es un **falso positivo**.
* Creemos que hay un efecto o diferencia, pero en realidad **no la hay**.

### 📌 Ejemplo:

Supón que una medicina **no mejora** la salud, pero en nuestro test de hipótesis concluimos que **sí mejora** → hemos cometido un error tipo I.

### ⚖️ Probabilidad de cometerlo:

* Se controla con el nivel de significancia $\alpha$ (ej. 0.05 = 5%).

---

## 🔵 **Error Tipo II (β)**

> **No rechazar la hipótesis nula (H₀) siendo falsa.**

### 🧠 Interpretación:

* Es un **falso negativo**.
* Creemos que **no hay diferencia**, cuando en realidad **sí la hay**.

### 📌 Ejemplo:

La medicina **sí mejora la salud**, pero el test dice que **no hay mejora significativa** → cometimos un error tipo II.

### 📊 ¿Cómo reducirlo?

* Aumentando el tamaño de la muestra.
* Aumentando el nivel de significancia.
* Disminuyendo la variabilidad de los datos.

---

## ⚖️ Tabla resumen

| Decisión           | H₀ verdadera       | H₀ falsa            |
| ------------------ | ------------------ | ------------------- |
| **No rechazar H₀** | ✅ Correcto         | ❌ **Error tipo II** |
| **Rechazar H₀**    | ❌ **Error tipo I** | ✅ Correcto          |

---

## 📈 Relación entre α, β y **potencia estadística**

* La **potencia** de una prueba es $1 - β$ → es la **probabilidad de detectar un efecto cuando realmente existe**.
* Existe un **trade-off** entre α y β: reducir uno puede aumentar el otro.
