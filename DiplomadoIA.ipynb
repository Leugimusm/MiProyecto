Cuando realizamos un test de hiptesis, siempre hay **incertidumbre**, ya que trabajamos con **muestras** y no con toda la poblacin. Por eso, podemos cometer dos tipos de errores:

##  **Error Tipo I ()**

> **Rechazar la hiptesis nula (H) siendo verdadera.**

###  Interpretacin:

* Es un **falso positivo**.
* Creemos que hay un efecto o diferencia, pero en realidad **no la hay**.

###  Ejemplo:

Supn que una medicina **no mejora** la salud, pero en nuestro test de hiptesis concluimos que **s mejora**  hemos cometido un error tipo I.

###  Probabilidad de cometerlo:

* Se controla con el nivel de significancia $\alpha$ (ej. 0.05 = 5%).

---

##  **Error Tipo II ()**

> **No rechazar la hiptesis nula (H) siendo falsa.**

###  Interpretacin:

* Es un **falso negativo**.
* Creemos que **no hay diferencia**, cuando en realidad **s la hay**.

###  Ejemplo:

La medicina **s mejora la salud**, pero el test dice que **no hay mejora significativa**  cometimos un error tipo II.

###  Cmo reducirlo?

* Aumentando el tamao de la muestra.
* Aumentando el nivel de significancia.
* Disminuyendo la variabilidad de los datos.
